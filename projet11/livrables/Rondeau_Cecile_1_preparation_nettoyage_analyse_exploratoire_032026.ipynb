```python
filepath = 'DisponibiliteAlimentaire_2017.csv'


# 2. Chargement
dispo_alimen_2017_df = pd.read_csv(filepath, sep=',', encoding='utf-8', low_memory=False)

# 3. Visualisation professionnelle
print("=" * 50)
print(f"SHAPE     : {dispo_alimen_2017_df.shape[0]} lignes × {dispo_alimen_2017_df.shape[1]} colonnes")
print("=" * 50)
print("\nCOLONNES & TYPES :")
print(dispo_alimen_2017_df.dtypes)
print("\nAPERÇU (5 premières lignes) :")
display(dispo_alimen_2017_df.head())
print("\nNaN PAR COLONNE :")
print(dispo_alimen_2017_df.isna().sum())
print(f"\nDOUBLONS : {dispo_alimen_2017_df.duplicated().sum()}")
```

```python
#️⃣ Identifier le bon produit: Viande de Volailles

print(sorted(dispo_alimen_2017_df['Produit'].unique()))
```

```python
#️⃣ Conserver le bon produit : Viande de Volailles 

dispo_volaille_df = dispo_alimen_2017_df[
    dispo_alimen_2017_df['Produit'] == 'Viande de Volailles'
].copy()

print("\nAPERÇU (5 premières lignes) :")
display(dispo_volaille_df.head())

print(sorted(dispo_volaille_df['Élément'].unique()))
print(sorted(dispo_volaille_df['Unité'].unique()))
```

```python
#️⃣ Filtrer le bon produit + les 3 éléments utiles + la bonne unité

dispo_volaille_df = dispo_alimen_2017_df[
    dispo_alimen_2017_df['Produit'] == 'Viande de Volailles'
].copy()

#️⃣ Filtrage des 4 éléments utiles (toutes unités confondues)
elements_utiles = [
    'Disponibilité intérieure',
    'Importations - Quantité',
    'Exportations - Quantité',
    'Disponibilité alimentaire en quantité (kg/personne/an)'
]

dispo_volaille_df = dispo_volaille_df[
    dispo_volaille_df['Élément'].isin(elements_utiles)
].copy()

#️⃣ Vérification
print(dispo_volaille_df.shape)
print(sorted(dispo_volaille_df['Élément'].unique()))
print(sorted(dispo_volaille_df['Unité'].unique()))
display(dispo_volaille_df.head(8))
```

```python
#️⃣ DISPO ALIMENTAIRE — Pivot : 1 ligne = 1 pays

#️⃣ Pivot sur les éléments
dispo_2017_df = dispo_volaille_df.pivot_table(
    index='Zone',
    columns='Élément',
    values='Valeur'
).reset_index()

#️⃣ Supprimer le nom parasite de l'axe columns
dispo_2017_df.columns.name = None

#️⃣ Renommer les colonnes
dispo_2017_df = dispo_2017_df.rename(columns={
    'Zone'                                              : 'pays',
    'Disponibilité intérieure'                          : 'dispo_interieure_kt',
    'Importations - Quantité'                           : 'importations_kt',
    'Exportations - Quantité'                           : 'exportations_kt',
    'Disponibilité alimentaire en quantité (kg/personne/an)' : 'conso_volaille_kg_hab_an'
})

#️⃣ Conversion milliers de tonnes → tonnes pour les 3 premières
for col in ['dispo_interieure_kt', 'importations_kt', 'exportations_kt']:
    dispo_2017_df[col] = dispo_2017_df[col] * 1000

#️⃣ Renommer après conversion
dispo_2017_df = dispo_2017_df.rename(columns={
    'dispo_interieure_kt' : 'dispo_interieure_tonnes',
    'importations_kt'     : 'importations_tonnes',
    'exportations_kt'     : 'exportations_tonnes'
})

#️⃣ Vérification

print(f"SHAPE : {dispo_2017_df.shape[0]} pays × {dispo_2017_df.shape[1]} colonnes")

print(f"\nNaN par colonne :\n{dispo_2017_df.isna().sum()}")
print(f"\nDoublons pays : {dispo_2017_df['pays'].duplicated().sum()}")
display(dispo_2017_df.head())
```

```python
# SOURCE 3 — Gouvernance World Bank - Correction : skiprows=4 pour sauter les métadonnées World Bank


filepath = 'worldbank_stabilite_politique_2017.csv'


# CORRECTION : skiprows=4 pour ignorer les 4 lignes de métadonnées
gouv_df = pd.read_csv(
    filepath,
    sep=',',
    encoding='utf-8',
    low_memory=False,
    skiprows=4          # ← ligne ajoutée
)

# Visualisation
print("=" * 50)
print(f"SHAPE     : {gouv_df.shape[0]} lignes × {gouv_df.shape[1]} colonnes")
print("=" * 50)
print("\nCOLONNES & TYPES :")
print(gouv_df.dtypes)
print("\nAPERÇU (5 premières lignes) :")
display(gouv_df.head())
print("\nNaN PAR COLONNE :")
print(gouv_df.isna().sum())
print(f"\nDOUBLONS : {gouv_df.duplicated().sum()}")
```

```python
#️⃣ ♦️ SOURCE 5 — Commerce extérieur poulet - faostat_Import_poulet_2017.csv

filepath = 'faostat_Import_poulet_2017.csv'

import_poulet_df = pd.read_csv(filepath, sep=',', encoding='utf-8', low_memory=False)

print("=" * 50)
print(f"SHAPE     : {import_poulet_df.shape[0]} lignes × {import_poulet_df.shape[1]} colonnes")
print("=" * 50)
print("\nCOLONNES & TYPES :")
print(import_poulet_df.dtypes)
print("\nÉLÉMENTS DISPONIBLES :")
print(sorted(import_poulet_df['Élément'].unique()))
print("\nPRODUITS DISPONIBLES :")
print(sorted(import_poulet_df['Produit'].unique()))
print("\nUNITÉS DISPONIBLES :")
print(sorted(import_poulet_df['Unité'].unique()))
print("\nAPERÇU (5 premières lignes) :")
display(import_poulet_df.head())
print("\nNaN PAR COLONNE :")
print(import_poulet_df.isna().sum())
print(f"\nDOUBLONS : {import_poulet_df.duplicated().sum()}")
```

```python
#️⃣ ♦️ SOURCE 5 — Commerce extérieur poulet — Nettoyage - 1 ligne = 1 pays | import_qty_tonnes | import_value_kusd

#️⃣ ♦️  Pivot : 1 ligne = 1 pays
import_piv = import_poulet_df.pivot_table(
    index='Zone',
    columns='Élément',
    values='Valeur'
).reset_index()

#️⃣ ♦️  Supprimer le nom parasite
import_piv.columns.name = None

#️⃣ ♦️  Renommer
import_poulet_2017_df = import_piv.rename(columns={
    'Zone'                      : 'pays',
    'Importations - quantité'   : 'import_qty_tonnes',
    'Importations - valeur'     : 'import_value_kusd'
})

#️⃣ ♦️  Créer le flag AVANT tout fillna
import_poulet_2017_df['import_declared'] = (
    import_poulet_2017_df['import_qty_tonnes'].notna().astype(int)
)

#️⃣ ♦️  NaN → 0 uniquement APRÈS le flag
import_poulet_2017_df['import_qty_tonnes'] = import_poulet_2017_df['import_qty_tonnes'].fillna(0)
import_poulet_2017_df['import_value_kusd'] = import_poulet_2017_df['import_value_kusd'].fillna(0)

#️⃣ ♦️  Vérification

print(f"SHAPE : {import_poulet_2017_df.shape[0]} pays × {import_poulet_2017_df.shape[1]} colonnes")

print(f"\nNaN par colonne :\n{import_poulet_2017_df.isna().sum()}")
print(f"\nDoublons pays : {import_poulet_2017_df['pays'].duplicated().sum()}")
print(f"\nPays avec import déclaré : {import_poulet_2017_df['import_declared'].sum()}")
print(f"Pays sans import déclaré : {(import_poulet_2017_df['import_declared'] == 0).sum()}")
print("\nAPERÇU :")
display(import_poulet_2017_df.head())
```

```python
#️⃣ ♦️ SCRIPT 2B — Mapping automatique pycountry - Objectif : trouver l'ISO3 pour chaque nom de pays

# Dictionnaire de correction manuelle — cas FAO non standards
# Ces noms échouent avec pycountry : on les corrige manuellement

corrections_manuelles = {
    # Noms avec parenthèses FAO
    "Bolivie (État plurinational de)":              "BOL",
    "Iran (République islamique d')":               "IRN",
    "Micronésie (États fédérés de)":               "FSM",
    "Venezuela (République bolivarienne du)":       "VEN",
    "Pays-Bas (Royaume des)":                       "NLD",
    # Noms longs FAO
    "Fédération de Russie":                         "RUS",
    "République démocratique du Congo":             "COD",
    "République populaire démocratique de Corée":   "PRK",
    "République de Corée":                          "KOR",
    "République centrafricaine":                    "CAF",
    "République dominicaine":                       "DOM",
    "République de Moldova":                        "MDA",
    "République arabe syrienne":                    "SYR",
    "République démocratique populaire lao":        "LAO",
    "République-Unie de Tanzanie":                  "TZA",
    # Chines
    "Chine, continentale":                          "CHN",
    "Chine — RAS de Hong-Kong":                     "HKG",
    "Chine — RAS de Macao":                         "MAC",
    "Chine, Taiwan Province de":                    "TWN",
    # Autres cas FAO
    "Royaume-Uni de Grande-Bretagne et d'Irlande du Nord": "GBR",
    "États-Unis d'Amérique":                        "USA",
    "Viet Nam":                                     "VNM",
    "Eswatini":                                     "SWZ",
    "Türkiye":                                      "TUR",
    "Cabo Verde":                                   "CPV",
    "Macédoine du Nord":                            "MKD",
    "Sahara occidental":                            "ESH",
    "Palestine":                                    "PSE",
    "Kosovo":                                       "XKX",
}

# Fonction de recherche ISO3 - Elle essaie pycountry d'abord, puis le dictionnaire manuel

def trouver_iso3(nom_pays):
    # Étape 1 : vérifier le dictionnaire manuel en priorité
    if nom_pays in corrections_manuelles:
        return corrections_manuelles[nom_pays]
    
    # Étape 2 : recherche exacte pycountry
    try:
        pays = pycountry.countries.lookup(nom_pays)
        return pays.alpha_3
    except LookupError:
        pass
    
    # Étape 3 : recherche floue pycountry (fuzzy)
    try:
        resultats = pycountry.countries.search_fuzzy(nom_pays)
        if resultats:
            return resultats[0].alpha_3
    except LookupError:
        pass
    
    # Étape 4 : échec total
    return None

# Application sur tous les noms
mapping_iso3 = {}

for nom in sorted(tous_les_pays):
    iso3 = trouver_iso3(nom)
    mapping_iso3[nom] = iso3

# Résumé
total         = len(mapping_iso3)
trouvés       = sum(1 for v in mapping_iso3.values() if v is not None)
non_trouvés   = sum(1 for v in mapping_iso3.values() if v is None)

print(f"Total        : {total}")
print(f"✅ Mappés    : {trouvés}")
print(f"❌ Échecs    : {non_trouvés}")

# Affichage des échecs uniquement
print("\n=== Noms sans ISO3 trouvé ===\n")
for nom, iso3 in sorted(mapping_iso3.items()):
    if iso3 is None:
        print(f"  ❌ {nom}")
```

```python
#️⃣ ♦️ SCRIPT 2D — Test country_converter corrigé - On filtre les retours qui ne sont pas des vrais codes ISO3

import country_converter as coco
import re

def est_iso3_valide(code):
    # Un vrai code ISO3 = exactement 3 lettres majuscules
    if code is None:
        return False
    return bool(re.match(r'^[A-Z]{3}$', str(code)))

cas_test = [
    "Afrique du Sud",
    "Allemagne",
    "Brésil",
    "Fédération de Russie",
    "Iran (République islamique d')",
    "Bolivie (État plurinational de)",
    "Chine, continentale",
    "Chine — RAS de Hong-Kong",
    "États-Unis d'Amérique",
    "Viet Nam",
    "République démocratique du Congo",
    "Côte d'Ivoire",
    "Allemagne",
    "Japon",
    "Singapour",
    "Turquie",
]

print("=== Test country_converter corrigé ===\n")
for nom in cas_test:
    iso3 = coco.convert(names=nom, to="ISO3", not_found=None)
    if not est_iso3_valide(iso3):
        iso3 = None
    statut = "✅" if iso3 else "❌"
    print(f"  {statut} {nom} → {iso3}")
```

1. Supprimer les pays avec trop de NaN sur les variables clés de l'ACP — un pays sans PIB, sans stabilité politique ET sans consommation n'est pas exploitable
2. Vérifier le seuil des 100 pays — exigence du projet

```python
#️⃣ ♦️ Filtre NaN sur variables clés ACP - Seuil : max 2 NaN tolérés sur les 7 variables clés

#️⃣ Variables clés retenues pour l'ACP
variables_cles = [
    "population",
    "pib_par_habitant_usd",
    "stabilite_politique",
    "conso_volaille_kg_hab_an",
    "import_qty_tonnes",
    "production_tonnes",
    "distance_km"
]

#️⃣ Comptage des NaN par pays sur ces 7 variables uniquement
base_df["nb_nan_cles"] = base_df[variables_cles].isnull().sum(axis=1)

#️⃣ Distribution des NaN pour décider
print("=== Distribution des NaN sur variables clés ===\n")
print(base_df["nb_nan_cles"].value_counts().sort_index())

print(f"\nPays avec 0 NaN  : {(base_df['nb_nan_cles'] == 0).sum()}")
print(f"Pays avec 1 NaN  : {(base_df['nb_nan_cles'] == 1).sum()}")
print(f"Pays avec 2 NaN  : {(base_df['nb_nan_cles'] == 2).sum()}")
print(f"Pays avec 3+ NaN : {(base_df['nb_nan_cles'] >= 3).sum()}")
print(f"\nTotal gardés (≤2 NaN) : {(base_df['nb_nan_cles'] <= 2).sum()}")
print(f"Total exclus (>2 NaN) : {(base_df['nb_nan_cles'] > 2).sum()}")
```

```python
#️⃣ ♦️ Filtre strict : 0 NaN sur variables clés ACP - Exigence auto-évaluation : aucun vide dans la matrice ACP

print(f"Pays avant filtrage : {base_df.shape[0]}")

#️⃣ On garde uniquement les pays complets sur les 7 variables clés
base_df = base_df[base_df["nb_nan_cles"] == 0].copy()

#️⃣ Suppression colonne de comptage
base_df = base_df.drop(columns=["nb_nan_cles"])

print(f"Pays après filtrage : {base_df.shape[0]}")
print(f"Colonnes            : {base_df.shape[1]}")
print(f"\n=== NaN par colonne ===")
print(base_df.isnull().sum())
```

```python
#️⃣♦️ Formule : consommation par habitant (kg) × population / 1000 - La division par 1000 convertit les kg en tonnes

base_df['marche_volaille_t'] = (
    base_df['conso_volaille_kg_hab_an'] * base_df['population'] / 1000
)

# Vérification 
print("NaN :", base_df['marche_volaille_t'].isna().sum())  # Attendu : 0
print(base_df['marche_volaille_t'].describe().round(0))
```

    📊 – Variable  2/3 : taux_dependance_import   -   Part des importations poulet dans la disponibilité intérieure volaille



</div>

```python
#️⃣♦️ Formule : imports poulet TCL / disponibilité intérieure volaille
     # Résultat attendu : majoritairement entre 0 et 1
     # Valeurs > 1 possibles — proxy imparfait entre deux sources différentes

base_df['taux_dependance_import'] = (
    base_df['import_qty_tonnes'] / base_df['dispo_interieure_tonnes']
)

#️⃣♦️ Vérification
print("NaN :", base_df['taux_dependance_import'].isna().sum())  # Attendu : 0
print(base_df['taux_dependance_import'].describe().round(3))

#️⃣♦️ Combien de pays ont un taux > 1 ? (signal à surveiller)
print("\nPays avec taux > 1 :",
      (base_df['taux_dependance_import'] > 1).sum())
```

<div style="background-color: #e8f5e9; padding: 12px; border-left: 5px solid #2e7d32; border-radius: 6px;">
    
    📊 – Variable  3/3 : taux_couverture_locale   -   Mesure si le pays couvre son marché avec sa production locale



</div>

```python
#️⃣♦️  Formule : production locale / disponibilité intérieure
# < 1 = le pays ne produit pas assez, il doit importer → cible potentielle
# > 1 = le pays est autosuffisant ou exportateur → cible moins prioritaire
base_df['taux_couverture_locale'] = (
    base_df['production_tonnes'] / base_df['dispo_interieure_tonnes']
)

#️⃣♦️  Vérification ---
print("NaN :", base_df['taux_couverture_locale'].isna().sum())  # Attendu : 0
print(base_df['taux_couverture_locale'].describe().round(3))

#️⃣♦️  Répartition métier
print("\nPays < 1 (besoin d'import) :",
      (base_df['taux_couverture_locale'] < 1).sum())
print("Pays > 1 (autosuffisants)  :",
      (base_df['taux_couverture_locale'] > 1).sum())
```
